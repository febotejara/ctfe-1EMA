#**Exemplo - Cálculo de variância e desvio padrão**

```python
n1, n2, n3, n4= input('insira os números da distribuição: ').split()
n1 = int(n1)
n2 = int(n2)
n3 = int(n3)
n4 = int(n4)
```

```python
# Média
media = (n1 + n2 + n3 + n4)/4
print(media)
```

```python
# Desvios
d1 = n1 - media
d2 = n2 - media
d3 = n3 - media
d4 = n4 - media
d1, d2, d3, d4
```

```python
# Somatória dos quadrados dos desvios
soma = d1**2 + d2**2 + d3**2 + d4**2
# Variância da distribuição (somatória dos elementos, dividido pela quantidade)
variancia = soma / 4
variancia
```

```python
# Desvio padrão
desvio_padrao = variancia**0.5
print(f'Desvio padrão = {desvio_padrao:.2f}')
```
