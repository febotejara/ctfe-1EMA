```python
temperatura = float(input())

#obrigado colocar a condição em "if" e "elif" e "else" NÃO TEM CONDIÇÃO NUNCA
if (temperatura > 30):
  print ('Temperatura alta :D')
elif (temperatura > 20):
  print ('Temperatura agradável :)')
else:
  print('Temperatura baixa :(')
```

```python
N1, N2, N3, N4= input('Insira os valores da distribuição: ').split()
N1 = float(N1)
N2 = float(N2)
N3 = float(N3)
N4 = float(N4)

media = (N1*2 + N2*3 + N3*4 + N4*1 ) / 10
if (media >= 7):
 print(f'Aluno aprovado:\n Média{media:.1f}!')
elif (media < 5):
 print(f'Aluno reprovado:\n Média{media:.1f}!')
else:
  print(f'Aluno de exame:\n Média{media:.1f}!')
```
